Baseline модель для прогнозирования остаточного ресурса (RUL) авиационных двигателей.
Данные: NASA CMAPSS (FD001).
Задача: регрессия, предсказание количества циклов до отказа.

```python
# @title 1. Установка библиотек

!pip install -q pandas numpy matplotlib seaborn scikit-learn xgboost mlflow joblib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from google.colab import files
import zipfile
import joblib
import mlflow
import mlflow.xgboost
import json
import os
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

```

```python
# @title 2. Загрузка данных

print("Загрузите файл CMAPSSData.zip")
uploaded = files.upload()
filename = list(uploaded.keys())[0]

with zipfile.ZipFile(filename, 'r') as zip_ref:
    zip_ref.extractall('/content/cmapss_data/')

print("Данные распакованы в /content/cmapss_data/")
```

```python
# @title 3. Загрузки датасета FD001

def load_cmapss(dataset='FD001'):
    columns = ['unit', 'cycle', 'setting1', 'setting2', 'setting3'] + \
              [f'sensor{i}' for i in range(1, 22)]

    train = pd.read_csv(f'/content/cmapss_data/train_{dataset}.txt',
                        sep='\s+', header=None, names=columns)
    test = pd.read_csv(f'/content/cmapss_data/test_{dataset}.txt',
                       sep='\s+', header=None, names=columns)
    rul = pd.read_csv(f'/content/cmapss_data/RUL_{dataset}.txt',
                      sep='\s+', header=None, names=['RUL'])

    train = train.dropna(axis=1, how='all')
    test = test.dropna(axis=1, how='all')

    return train, test, rul

train_fd001, test_fd001, rul_fd001 = load_cmapss('FD001')
```

```python
# @title 5. Создание целевой переменной RUL

def add_rul_column(df):
    """
    Добавляет колонку RUL = максимальный цикл - текущий цикл.
    Для каждого двигателя отдельно.
    """
    df = df.copy()
    max_cycles = df.groupby('unit')['cycle'].transform('max')
    df['RUL'] = max_cycles - df['cycle']
    return df

train_with_rul = add_rul_column(train_fd001)


print("ЦЕЛЕВАЯ ПЕРЕМЕННАЯ RUL")
print("\nПервые 10 строк с добавленной колонкой RUL:")
print(train_with_rul[['unit', 'cycle', 'RUL']].head(10))

print("\nСтатистика целевой переменной RUL:")
print(f"  Минимум: {train_with_rul['RUL'].min()}")
print(f"  Максимум: {train_with_rul['RUL'].max()}")
print(f"  Среднее: {train_with_rul['RUL'].mean():.1f}")
print(f"  Медиана: {train_with_rul['RUL'].median()}")

# Визуализация распределения RUL
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.hist(train_with_rul['RUL'], bins=50, edgecolor='black', alpha=0.7)
plt.xlabel('RUL (циклы)')
plt.ylabel('Частота')
plt.title('Распределение RUL')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
for unit in range(1, 6):
    unit_data = train_with_rul[train_with_rul['unit'] == unit]
    plt.plot(unit_data['cycle'], unit_data['RUL'], label=f'Engine {unit}')
plt.xlabel('Цикл')
plt.ylabel('RUL')
plt.title('Деградация двигателей (первые 5)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# @title 6. Отбор признаков

sensor_cols = [col for col in train_with_rul.columns if col.startswith('sensor')]

"""
Считаем коэффициент корреляции Пирсона с целевой переменной RUL
Исключаем сенсоры с постоянными значениями (корреляция nan)
"""

correlations = {}
for sensor in sensor_cols:
    corr = train_with_rul[sensor].corr(train_with_rul['RUL'])
    # Пропускаем сенсоры с nan корреляцией (постоянные значения, не несут информации)
    if not pd.isna(corr):
        correlations[sensor] = corr

# Преобразуем в список пар и сортируем
corr_list = [(sensor, corr) for sensor, corr in correlations.items()]
corr_list.sort(key=lambda x: abs(x[1]), reverse=True)

print("Топ-10 сенсоров по корреляции с RUL:")
for i, (sensor, corr) in enumerate(corr_list[:10]):
    print(f"  {i+1}. {sensor}: {corr:.4f}")

# Берем топ-5 сенсоров как простой список строк
top_sensors = [sensor for sensor, corr in corr_list[:5]]
print(f"\nОтобрано сенсоров: {top_sensors}")

# Проверяем, что это строки
print(f"Тип первого элемента: {type(top_sensors[0])}")
print(f"Значение: {top_sensors[0]}")

# Создаем DataFrame для визуализации
corr_df = pd.DataFrame(corr_list[:10], columns=['sensor', 'correlation'])

# Визуализация корреляций
plt.figure(figsize=(10, 6))
colors = ['green' if x > 0 else 'red' for x in corr_df['correlation']]
plt.barh(corr_df['sensor'], corr_df['correlation'], color=colors)
plt.xlabel('Коэффициент корреляции')
plt.title('Корреляция сенсоров с остаточным ресурсом (RUL)')
plt.gca().invert_yaxis()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# @title 7. FEATURE ENGINEERING: скользящие окна

def add_rolling_features(df, sensors, windows=[5, 10]):
    df = df.copy()

    for sensor in sensors:
        print(f"Обработка: {sensor}")

        for window in windows:
            col_name = f'{sensor}_rolling_mean_{window}'
            # Вычисляем скользящее среднее для каждого двигателя
            df[col_name] = df.groupby('unit')[sensor].rolling(window, min_periods=1).mean().reset_index(level=0, drop=True)

    return df

# Добавляем новые колонки в данных
train_with_features = add_rolling_features(train_with_rul, top_sensors, windows=[5, 10])

print(f"\nРазмер данных после добавления колонок: {train_with_features.shape}")

# Список признаков
base_features = top_sensors + ['setting1', 'setting2', 'setting3']
rolling_features = [col for col in train_with_features.columns if 'rolling' in col]
selected_features = base_features + rolling_features

print(f"Итого")
print(f"\nВсего признаков: {len(selected_features)}")
print(f"Базовые: {base_features}")
print(f"Скользящие средние: {rolling_features[:5]}...")
```

```python
# @title 8. Подготовка данных для обучения

# Нормализация и разделение на train/val

X = train_with_features[selected_features]
y = train_with_features['RUL']

# Удаляем строки с NaN
X = X.dropna()
y = y.loc[X.index]

# Нормализация
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Разделение на обучающую и валидационную выборки
X_train, X_val, y_train, y_val = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
print(f"Количество признаков: {X.shape[1]}")
print(f"Обучающая выборка: {X_train.shape[0]} строк")
print(f"Валидационная выборка: {X_val.shape[0]} строк")
```

```python
# @title 9. Обучение модели регресии

# Используем XGBoost как базовый алгоритм

model = xgb.XGBRegressor(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.1,
    random_state=42
)


print("Обучение модели")
print("Параметры: n_estimators=200, max_depth=6, learning_rate=0.1")

model.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=False)

y_pred_train = model.predict(X_train)
y_pred_val = model.predict(X_val)

train_rmse = np.sqrt(mean_squared_error(y_train, y_pred_train))
val_rmse = np.sqrt(mean_squared_error(y_val, y_pred_val))
train_mae = mean_absolute_error(y_train, y_pred_train)
val_mae = mean_absolute_error(y_val, y_pred_val)
train_r2 = r2_score(y_train, y_pred_train)
val_r2 = r2_score(y_val, y_pred_val)

print("\nРезультаты:")
print(f"  Обучающая RMSE: {train_rmse:.2f}")
print(f"  Валидационная RMSE: {val_rmse:.2f}")
print(f"  Обучающая MAE: {train_mae:.2f}")
print(f"  Валидационная MAE: {val_mae:.2f}")
print(f"  Обучающая R²: {train_r2:.3f}")
print(f"  Валидационная R²: {val_r2:.3f}")
```

```python
# @title 10. Сравнение: топ-5 сенсоров vs все 21 сенсор


print("СРАВНЕНИЕ: ТОП-5 СЕНСОРОВ vs ВСЕ 21 СЕНСОР")

#1. Модель на топ-5 сенсорах (уже обучена в ячейке 9)

print("\n1. Модель на топ-5 сенсорах + скользящие окна")
print(f"  MAE: {val_mae:.2f}")
print(f"  RMSE: {val_rmse:.2f}")
print(f"  R²: {val_r2:.3f}")

#2. Модель на всех 21 сенсоре + скользящие окна

print("\n2. Модель на всех 21 сенсоре + скользящие окна")

# Получаем все сенсоры (исключая постоянные)
all_sensors = [sensor for sensor, corr in correlations.items()]
print(f"   Всего сенсоров (после фильтрации): {len(all_sensors)}")

# Добавляем скользящие окна для всех сенсоров
train_with_features_all = add_rolling_features(train_with_rul, all_sensors, windows=[5, 10])

# Список признаков
base_features_all = all_sensors + ['setting1', 'setting2', 'setting3']
rolling_features_all = [col for col in train_with_features_all.columns if 'rolling' in col]
selected_features_all = base_features_all + rolling_features_all

print(f"   Всего признаков: {len(selected_features_all)}")

# Подготовка данных
X_all = train_with_features_all[selected_features_all]
y_all = train_with_features_all['RUL']

X_all = X_all.dropna()
y_all = y_all.loc[X_all.index]

scaler_all = StandardScaler()
X_all_scaled = scaler_all.fit_transform(X_all)

X_train_all, X_val_all, y_train_all, y_val_all = train_test_split(
    X_all_scaled, y_all, test_size=0.2, random_state=42
)

# Обучение
model_all = xgb.XGBRegressor(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.1,
    random_state=42
)

model_all.fit(X_train_all, y_train_all, eval_set=[(X_val_all, y_val_all)], verbose=False)

y_pred_val_all = model_all.predict(X_val_all)

all_mae = mean_absolute_error(y_val_all, y_pred_val_all)
all_rmse = np.sqrt(mean_squared_error(y_val_all, y_pred_val_all))
all_r2 = r2_score(y_val_all, y_pred_val_all)

print(f"\nРезультаты модели на всех 21 сенсоре:")
print(f"  MAE: {all_mae:.2f}")
print(f"  RMSE: {all_rmse:.2f}")
print(f"  R²: {all_r2:.3f}")

#3. Сравнение
print(f"Топ-5 сенсоров ({len(selected_features)} признаков):   MAE = {val_mae:.2f}, R² = {val_r2:.3f}")
print(f"Все 21 сенсор ({len(selected_features_all)} признаков): MAE = {all_mae:.2f}, R² = {all_r2:.3f}")

if all_mae < val_mae:
    improvement = ((val_mae - all_mae) / val_mae) * 100
    print(f"\nМодель на всех сенсорах лучше на {improvement:.1f}%")
else:
    improvement = ((all_mae - val_mae) / all_mae) * 100
    print(f"\nМодель на топ-5 сенсорах лучше на {improvement:.1f}%")
```

```python
# @title 11. Визуализация результатов

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# График 1: предсказанные vs фактические значения
axes[0, 0].scatter(y_val_all, y_pred_val_all, alpha=0.3, s=10)
axes[0, 0].plot([y_val_all.min(), y_val_all.max()], [y_val_all.min(), y_val_all.max()], 'r--', lw=2)
axes[0, 0].set_xlabel('Фактический RUL (циклы)')
axes[0, 0].set_ylabel('Предсказанный RUL (циклы)')
axes[0, 0].set_title(f'Предсказанные vs фактические значения\nRMSE: {all_rmse:.2f}, R²: {all_r2:.3f}')
axes[0, 0].grid(True, alpha=0.3)

# График 2: распределение ошибок
errors = y_val_all - y_pred_val_all
axes[0, 1].hist(errors, bins=50, edgecolor='black', alpha=0.7)
axes[0, 1].axvline(x=0, color='red', linestyle='--')
axes[0, 1].set_xlabel('Ошибка предсказания (циклы)')
axes[0, 1].set_ylabel('Частота')
axes[0, 1].set_title(f'Распределение ошибок\nMAE: {all_mae:.2f}')
axes[0, 1].grid(True, alpha=0.3)

# График 3: важность признаков (топ-20)
importance_df = pd.DataFrame({
    'feature': selected_features_all,
    'importance': model_all.feature_importances_
}).sort_values('importance', ascending=True).tail(20)

axes[1, 0].barh(importance_df['feature'], importance_df['importance'])
axes[1, 0].set_xlabel('Важность')
axes[1, 0].set_title('Топ-20 важных признаков (XGBoost)')
axes[1, 0].grid(True, alpha=0.3)

# График 4: кривая обучения
results = model_all.evals_result()
epochs = len(results['validation_0']['rmse'])
axes[1, 1].plot(range(epochs), results['validation_0']['rmse'], label='Валидация')
axes[1, 1].set_xlabel('Итерация')
axes[1, 1].set_ylabel('RMSE')
axes[1, 1].set_title('Кривая обучения')
axes[1, 1].legend()
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# @title 12. СОХРАНЕНИЕ МОДЕЛЕЙ И АРТЕФАКТОВ

os.makedirs('/content/models', exist_ok=True)

# Сохраняем модель на всех сенсорах
model_all.save_model('/content/models/xgb_rul_all.json')
joblib.dump(scaler_all, '/content/models/scaler_all.pkl')

with open('/content/models/features_all.json', 'w') as f:
    json.dump(selected_features_all, f)

# Сохраняем метрики
metrics = {
    'model_type': 'XGBoost_all_sensors',
    'val_rmse': all_rmse,
    'val_mae': all_mae,
    'val_r2': all_r2,
    'features_count': len(selected_features_all),
    'n_estimators': 200,
    'max_depth': 6,
    'learning_rate': 0.1,
    'rolling_windows': [5, 10]
}

with open('/content/models/metrics_all.json', 'w') as f:
    json.dump(metrics, f, indent=2)


print("СОХРАНЕНИЕ МОДЕЛИ НА ВСЕХ СЕНСОРАХ")
print("Модель и артефакты сохранены в /content/models/")
print(f"  val_mae: {all_mae:.2f}")
print(f"  val_rmse: {all_rmse:.2f}")
print(f"  val_r2: {all_r2:.3f}")
print(f"  features_count: {len(selected_features_all)}")

# MLflow логирование
mlflow.set_experiment("CMAPSS_Predictive_Maintenance")

with mlflow.start_run(run_name="All_Sensors_With_Rolling"):
    mlflow.log_params({
        "model_type": "XGBoost",
        "n_estimators": 200,
        "max_depth": 6,
        "learning_rate": 0.1,
        "features_count": len(selected_features_all),
        "rolling_windows": [5, 10]
    })
    mlflow.log_metrics({
        "val_rmse": all_rmse,
        "val_mae": all_mae,
        "val_r2": all_r2
    })
    mlflow.xgboost.log_model(model_all, "rul_regressor_all_sensors")

    print(f"\nЭксперимент залогирован в MLflow")
    print(f"Run ID: {mlflow.active_run().info.run_id}")

# Создаем архив и скачиваем
!zip -r /content/models_all.zip /content/models/ > /dev/null
files.download('/content/models_all.zip')

print("\nАрхив models_all.zip скачан на компьютер")
print("\nФайлы в архиве:")
print("  - xgb_rul_all.json (модель)")
print("  - scaler_all.pkl (нормализатор)")
print("  - features_all.json (список признаков)")
print("  - metrics_all.json (метрики)")
```

```python
# @title 13 Финальная оценка на test_fd001


print("ФИНАЛЬНАЯ ОЦЕНКА МОДЕЛИ НА TEST_FD001 (ВСЕ СЕНСОРЫ)")

def prepare_test_data_all(test_df, all_sensors, scaler_all, selected_features_all, windows=[5, 10]):
    """Подготавливает тестовые данные для модели на всех сенсорах"""
    test_with_features = add_rolling_features(test_df, all_sensors, windows=windows)
    X_test = test_with_features[selected_features_all]
    X_test = X_test.dropna()
    X_test_scaled = scaler_all.transform(X_test)
    return X_test_scaled, test_with_features

# Подготавливаем данные
X_test_scaled_all, test_with_features_all = prepare_test_data_all(
    test_fd001,
    all_sensors,
    scaler_all,
    selected_features_all,
    windows=[5, 10]
)

# Берем последнюю строку для каждого двигателя
last_cycles_all = test_with_features_all.groupby('unit').last().reset_index()
last_cycles_indices_all = last_cycles_all.index

# Предсказания для последних циклов
X_test_last_all = X_test_scaled_all[last_cycles_indices_all]
y_test_pred_all = model_all.predict(X_test_last_all)

# Правильные ответы
y_test_true = rul_fd001['RUL'].values

print(f"Количество предсказаний: {len(y_test_pred_all)}")
print(f"Количество правильных ответов: {len(y_test_true)}")
print(f"\nПервые 5 предсказаний: {y_test_pred_all[:5]}")
print(f"Первые 5 правильных: {y_test_true[:5]}")

# Метрики
test_mae_all = mean_absolute_error(y_test_true, y_test_pred_all)
test_rmse_all = np.sqrt(mean_squared_error(y_test_true, y_test_pred_all))
test_r2_all = r2_score(y_test_true, y_test_pred_all)

print(f"\nРезультаты на test_fd001 (модель на всех сенсорах):")
print(f"  MAE: {test_mae_all:.2f} циклов")
print(f"  RMSE: {test_rmse_all:.2f} циклов")
print(f"  R²: {test_r2_all:.3f}")

# Сравнение с предыдущей моделью
print("СРАВНЕНИЕ С МОДЕЛЬЮ НА ТОП-5 СЕНСОРАХ")
print(f"Топ-5 сенсоров:   MAE = {test_mae:.2f}, R² = {test_r2:.3f}")
print(f"Все сенсоры:      MAE = {test_mae_all:.2f}, R² = {test_r2_all:.3f}")

if test_mae_all < test_mae:
    improvement = ((test_mae - test_mae_all) / test_mae) * 100
    print(f"\nМодель на всех сенсорах лучше на {improvement:.1f}%")
else:
    improvement = ((test_mae_all - test_mae) / test_mae_all) * 100
    print(f"\nМодель на топ-5 сенсорах лучше на {improvement:.1f}%")
```
