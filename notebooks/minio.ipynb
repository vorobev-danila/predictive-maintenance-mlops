```python
import os
import boto3
from botocore.exceptions import ClientError

# Параметры MinIO
minio_endpoint = "192.168.0.104:9000"  # IP вашей VM + порт MinIO
access_key = "admin"              # по умолчанию MinIO
secret_key = "password123"              # по умолчанию MinIO
bucket_name = "mlops-predictive-maintenance-data"  # путь к бакету
```

```python
s3 = boto3.client(
    "s3",
    endpoint_url=f"http://{minio_endpoint}",
    aws_access_key_id=access_key,
    aws_secret_access_key=secret_key,
)
```

```python
bucket_name_simple = "mlops-predictive-maintenance-data"

# Создаём бакет, если не существует
try:
    s3.create_bucket(Bucket=bucket_name_simple)
    print(f"Bucket '{bucket_name_simple}' создан.")
except ClientError as e:
    if e.response['Error']['Code'] == 'BucketAlreadyOwnedByYou':
        print(f"Bucket '{bucket_name_simple}' уже существует.")
    else:
        raise
```

```python
s3.list_buckets()
```

```python
local_folder = r"C:\Users\user\Desktop\MLOps\predictive-maintenance-mlops\data\raw"

# --- префикс в S3 (это "папка") ---
s3_prefix = "raw/"

# --- загрузка файлов ---
for root, dirs, files in os.walk(local_folder):
    for file in files:
        local_path = os.path.join(root, file)
        
        # относительный путь (чтобы сохранить структуру)
        relative_path = os.path.relpath(local_path, local_folder)
        
        # ключ в S3
        s3_key = os.path.join(s3_prefix, relative_path).replace("\\", "/")
        
        # загрузка
        s3.upload_file(local_path, bucket_name, s3_key)
        
        print(f"Загружен: {s3_key}")
```
